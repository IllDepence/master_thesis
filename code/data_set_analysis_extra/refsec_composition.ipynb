````python
%matplotlib inline
import json
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

with open('citation_count_differences.json') as f:
    citcount = json.load(f)

citcount = sorted(citcount, key=lambda t: (-t[1], (t[0]+t[2])))
````

````python
only_unarxive = [t[0] for t in citcount]
both = [t[1] for t in citcount]
only_mag = [t[2] for t in citcount]
````

````python
plt.cla()
X = np.arange(0, len(citcount))
axes = plt.gca()
axes.set_ylim(-50, 50)
axes.set_aspect(3000)
plt.stackplot(X, *[only_unarxive, both, only_mag], baseline='wiggle', labels=['unarXive only', 'both', 'MAG only'])
plt.xlabel('document ID')
plt.ylabel('composition of reference\nsection coverage')
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1))
plt.draw()
f=plt.gcf()                               # https://github.com/matplotlib/matplotlib/issues/3139#issuecomment-46241351
f.set_size_inches(531./f.dpi,153./f.dpi)  # meh
plt.savefig('wigglecitations.svg', dpi=300)
````
